# This is the Matplotlib module of the deeplearning prereq stack

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0 , 20, 1000) # creates a 1d array of 1000 points between 0 and 20
```

```python
y = np.sin(x) + 0.2 * x
```

```python
plt.plot(x, y)
plt.xlabel("input")
plt.ylabel("output")
```

```python
plt.show()
```
